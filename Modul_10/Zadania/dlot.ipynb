```python
values_list = dataset.values.ravel()
print(values_list)
```

```python
dane = dict(Counter(values_list))
print(dane)
```

# Wyliczenia

```python
# Definiowanie przedziałów
przedzialy = {
    '1-7': 0,
    '8-14': 0,
    '15-21': 0,
    '22-28': 0,
    '29-35': 0,
    '36-42': 0,
    '43-49': 0
}

# Przypisz wartości wystąpień do odpowiednich przedziałów
for liczba, wystapienia in dane.items():
    liczba = int(liczba)  # Zamień klucz na liczbę całkowitą
    if 1 <= liczba <= 7:
        przedzialy['1-7'] += wystapienia
    elif 8 <= liczba <= 14:
        przedzialy['8-14'] += wystapienia
    elif 15 <= liczba <= 21:
        przedzialy['15-21'] += wystapienia
    elif 22 <= liczba <= 28:
        przedzialy['22-28'] += wystapienia
    elif 29 <= liczba <= 35:
        przedzialy['29-35'] += wystapienia
    elif 36 <= liczba <= 42:
        przedzialy['36-42'] += wystapienia
    elif 43 <= liczba <= 49:
        przedzialy['43-49'] += wystapienia

# Wypisz wyniki
for przedzial, dane_w_przedziale in przedzialy.items():
    print(f"Przedział {przedzial}: {dane_w_przedziale} wystąpień.")
```

```python
# Sortowanie przedziałów malejąco według liczby wystąpień
posortowane_przedzialy = sorted(przedzialy.items(), key=lambda x: x[1], reverse=True)

# Wyświetlanie 2 wartości, które wystąpiły najczęściej w każdym przedziale
for przedzial, dane in posortowane_przedzialy:
    print(f"Przedział {przedzial}: {dane_w_przedziale} wystąpień.")
    
# Tworzenie słownika wartości wystąpień dla danego przedziału
wartosci_wystapien = {}

for przedzial, dane_w_przedziale in posortowane_przedzialy:
    przedzial_start, przedzial_end = map(int, przedzial.split('-'))
    
    # Tworzenie listy wartości w danym przedziale
    wartosci_w_przedziale = [liczba for liczba in values_list if przedzial_start <= int(liczba) <= przedzial_end]
    
    # Tworzenie słownika wartości wystąpień w danym przedziale
    dane_w_przedziale = dict(Counter(wartosci_w_przedziale))
    
    # Sortowanie wartości wystąpień w danym przedziale
    posortowane_wartosci = sorted(dane_w_przedziale.items(), key=lambda x: x[1], reverse=True)
    
    # Wybieranie dwóch najczęściej występujących wartości
    wybrane_wartosci = posortowane_wartosci[:7]
    
    # Dodawanie wybranych wartości do słownika wynikowego
    wartosci_wystapien[przedzial] = wybrane_wartosci

# Wyświetlanie wybranych wartości
for przedzial, wybrane_wartosci in wartosci_wystapien.items():
    print(f"Przedział {przedzial}:")
    for wartosc, wystapienia in wybrane_wartosci:
        print(f" - Wartość: {wartosc}, wystąpienia: {wystapienia}")
```
