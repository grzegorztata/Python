```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler

x_100 = 3
sigma_100 = 1
data_100 = np.random.normal(x_100, sigma_100, 100).round(2)

x_10000 = 3
sigma_10000 = 1
data_10000 = np.random.normal(x_10000, sigma_10000, 10000).round(2)

print("Dane dla 100 próbek:")
print(data_100)

print("\nDane dla 10 000 próbek:")
print(data_10000)
```

```python
data_10000.min()
```

# Histogram

```python
plt.hist(data_100, bins=20)
plt.show()
```

```python
plt.hist(data_10000, bins=20)
plt.show()
```

###### Im większa liczba danych tym większe prawdopodobieństwo, że wykres Histogramowy będzie miał największą liczbę wartości zlokalizowanych bliżej średniej

# Średnia

```python
mean_100 = np.round(np.mean(data_100))
mean_100
```

```python
mean_10000 = np.round(np.mean(data_10000))
mean_10000
```

```python
plt.hist(data_100, bins=20)
plt.axvline(x=mean_100, color='red', label='Średnia')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(data_10000, bins=20)
plt.axvline(x=mean_10000, color='red', label='Średnia')
plt.legend(loc='upper right')
plt.show()
```

###### Tutaj zgodnie z zadaniem średnia wynosi 3.0

# Mediana

```python
median_100 = np.median(data_100)
median_100
```

```python
median_10000 = np.median(data_10000)
median_10000
```

```python
plt.hist(data_100, bins=20)
plt.axvline(x=mean_100,color='red',label='Średnia')
plt.axvline(x=median_100,color='green',label='Mediana')
plt.legend(loc='upper right')
plt.show
```

```python
plt.hist(data_10000, bins=20)
plt.axvline(x=mean_10000,color='red',label='Średnia')
plt.axvline(x=median_10000,color='green',label='Mediana')
plt.legend(loc='upper right')
plt.show
```

###### Im większa próba, tym większe prawdopodobieństwo , że Średnia i Mediana będą się na siebie nakładały (Średnia 3.0, Mediana 2.99)

# Moda

```python
mode_100 = stats.mode(data_100, keepdims=True)
mode_100
```

```python
mode_10000 = stats.mode(data_10000, keepdims=True)
mode_10000
```

```python
plt.hist(data_100, bins=20)
plt.axvline(x=mean_100,color='red',label='Średnia')
plt.axvline(x=mode_100[0],color='orange',label='Moda')
plt.legend(loc='upper right')
plt.show
```

```python
plt.hist(data_10000, bins=20)
plt.axvline(x=mean_10000,color='red',label='Średnia')
plt.axvline(x=mode_10000[0],color='orange',label='Moda')
plt.legend(loc='upper right')
plt.show
```

###### Im większa próba, tym większe prawdopodobieństwo , że Moda będzie się znajdowała bliżej Średniej

# Kwartyle

```python
q0_100 = np.quantile(data_100, 0.0)
q1_100 = np.quantile(data_100, 0.25)
q3_100 = np.quantile(data_100, 0.75)
q4_100 = np.quantile(data_100, 1.0)
print(f'Q0: {q0_100}')
print(f'Q1: {q1_100}')
print(f'Q2: {median_100}')
print(f'Q3: {q3_100}')
print(f'Q4: {q4_100}')
```

```python
q0_10000 = np.quantile(data_10000, 0.0)
q1_10000 = np.quantile(data_10000, 0.25)
q3_10000 = np.quantile(data_10000, 0.75)
q4_10000 = np.quantile(data_10000, 1.0)
print(f'Q0: {q0_10000}')
print(f'Q1: {q1_10000}')
print(f'Q2: {median_10000}')
print(f'Q3: {q3_10000}')
print(f'Q4: {q4_10000}')
```

```python
plt.hist(data_100, bins=20)
plt.axvline(x = q0_100, color='black', label = 'Min')
plt.axvline(x = q1_100, color='red', label = 'Q1')
plt.axvline(x = q3_100, color='blue', label = 'Q3')
plt.axvline(x = q4_100, color='green', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(data_10000, bins=20)
plt.axvline(x = q0_10000, color='black', label = 'Min')
plt.axvline(x = q1_10000, color='red', label = 'Q1')
plt.axvline(x = q3_10000, color='blue', label = 'Q3')
plt.axvline(x = q4_10000, color='green', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

###### Tutaj jedyne co mi przychodzi do głowy, to wartość ujemna dla próbki data_10000. Czy tutaj przypadkiem nie chodzi o to , że im większa próba tym większe prawdopodobieństwo na wystąpienie błędów?

# Wariancja

```python
variance_100 = np.var(data_100, ddof=1)
variance_100
```

```python
variance_10000 = np.var(data_10000, ddof=1)
variance_10000
```

###### Dla próbki variance_100 wariancja wyniosła 1.35, dla próbki variance_10000 wariancja wyniosła 0.99. Generalnie dla większej próbki wartości rozproszenia powinny być większe. Ale nie jest to żaden błąd,. Może to wynikać z różnic w samej naturze próbek danych. Przy kilkukrotnym wygenerowaniu danych wariancja była częściej większa dla próbki 'variance_10000' (stosunek 8:2)

# Odchylenie standardowe

```python
print(variance_100**(1/2))
print(variance_10000**(1/2))
```

###### Skoro Odchylenie Standardowe jest pierwiastkiem Wariancji to można oczekiwać, że dla próbki 100 wartości będzie ono w większości przypadków mniejsze niż dla próbki 10000 wartości

# Skalowanie zmiennej - StandardScaller

```python
scaler_standardized = StandardScaler()
standardized_data_100 = scaler_standardized.fit_transform(data_100.reshape(-1, 1))
plt.hist(standardized_data_100, bins=20)
plt.axvline(x = standardized_data_100.mean(), color='red', label = 'Średnia')
plt.show()
print(f'Średnia: {standardized_data_100.mean()}')
print(f'Odchylenie standardowe: {np.std(standardized_data_100, ddof=1)}')
```

```python
standardized_data_10000 = scaler_standardized.fit_transform(data_10000.reshape(-1, 1))
plt.hist(standardized_data_10000, bins=20)
plt.axvline(x = standardized_data_10000.mean(), color='red', label = 'Średnia')
plt.show()
print(f'Średnia: {standardized_data_10000.mean()}')
print(f'Odchylenie standardowe: {np.std(standardized_data_10000, ddof=1)}')
```

# Skalowanie zmiennej - MinMax Scaller

```python
scaler_MinMax = MinMaxScaler()
normalized_data_100 = scaler_MinMax.fit_transform(data_100.reshape(-1, 1))
plt.hist(normalized_data_100, bins=20)
plt.axvline(x = normalized_data_100.mean(), color='red', label = 'Średnia')
plt.show()
print(f'Średnia: {normalized_data_100.mean()}')
print(f'Odchylenie standardowe: {np.std(normalized_data_100, ddof=1)}')
```

```python
scaler_MinMax = MinMaxScaler()
normalized_data_10000 = scaler_MinMax.fit_transform(data_10000.reshape(-1, 1))
plt.hist(normalized_data_10000, bins=20)
plt.axvline(x = normalized_data_10000.mean(), color='red', label = 'Średnia')
plt.show()
print(f'Średnia: {normalized_data_10000.mean()}')
print(f'Odchylenie standardowe: {np.std(normalized_data_10000, ddof=1)}')
```

### Dla większej próby wskaźniki Statystyki opisowej są bardziej zbliżone do średniej. Nie jest to jednak żadną regułą i jeżeli tak się nie stanie, to nie powinniśmy tego traktować jako błąd. Może to wynikać z różnic w samej naturze próbek danych
