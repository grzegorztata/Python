```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from collections import Counter
```

```python
dataset = pd.read_csv('../Files/multi_multi.csv', sep=',', header=None)
dataset.head()
```

```python
columns_to_drop = [0, 1, 2, 23]
dataset = dataset.drop(columns = columns_to_drop, axis = 1)
dataset.head()
```

```python
dataset = dataset.tail(20)
dataset
```

```python
values_list = dataset.values.ravel()
print(values_list)
```

```python
dane = dict(Counter(values_list))
print(dane)
```

# Wyliczenia

```python
# Definiowanie przedziaÅ‚Ã³w
przedzialy = {
    '1-10': 0,
    '11-20': 0,
    '21-30': 0,
    '31-40': 0,
    '41-50': 0,
    '51-60': 0,
    '61-70': 0,
    '71-80': 0

}

# Przypisz wartoÅ›ci wystÄ…pieÅ„ do odpowiednich przedziaÅ‚Ã³w
for liczba, wystapienia in dane.items():
    liczba = int(liczba)  # ZamieÅ„ klucz na liczbÄ™ caÅ‚kowitÄ…
    if 1 <= liczba <= 10:
        przedzialy['1-10'] += wystapienia
    elif 11 <= liczba <= 20:
        przedzialy['11-20'] += wystapienia
    elif 21 <= liczba <= 30:
        przedzialy['21-30'] += wystapienia
    elif 31 <= liczba <= 40:
        przedzialy['31-40'] += wystapienia
    elif 41 <= liczba <= 50:
        przedzialy['41-50'] += wystapienia
    elif 51 <= liczba <= 60:
        przedzialy['51-60'] += wystapienia
    elif 61 <= liczba <= 70:
        przedzialy['61-70'] += wystapienia
    elif 71 <= liczba <= 80:
        przedzialy['71-80'] += wystapienia

# Wypisz wyniki
for przedzial, dane_w_przedziale in przedzialy.items():
    print(f"PrzedziaÅ‚ {przedzial}: {dane_w_przedziale} wystÄ…pieÅ„.")
```

```python
# # Sortowanie przedziaÅ‚Ã³w malejÄ…co wedÅ‚ug liczby wystÄ…pieÅ„
# posortowane_przedzialy = sorted(przedzialy.items(), key=lambda x: x[1], reverse=True)

# # WyÅ›wietlanie 2 wartoÅ›ci, ktÃ³re wystÄ…piÅ‚y najczÄ™Å›ciej w kaÅ¼dym przedziale
# for przedzial, dane in posortowane_przedzialy:
#     print(f"PrzedziaÅ‚ {przedzial}: {dane_w_przedziale} wystÄ…pieÅ„.")
    
# Tworzenie sÅ‚ownika wartoÅ›ci wystÄ…pieÅ„ dla danego przedziaÅ‚u
wartosci_wystapien = {}

for przedzial, dane_w_przedziale in posortowane_przedzialy:
    przedzial_start, przedzial_end = map(int, przedzial.split('-'))
    
    # Tworzenie listy wartoÅ›ci w danym przedziale
    wartosci_w_przedziale = [liczba for liczba in values_list if przedzial_start <= int(liczba) <= przedzial_end]
    
    # Tworzenie sÅ‚ownika wartoÅ›ci wystÄ…pieÅ„ w danym przedziale
    dane_w_przedziale = dict(Counter(wartosci_w_przedziale))
    
    # Sortowanie wartoÅ›ci wystÄ…pieÅ„ w danym przedziale
    posortowane_wartosci = sorted(dane_w_przedziale.items(), key=lambda x: x[1], reverse=True)
    
    # Wybieranie dwÃ³ch najczÄ™Å›ciej wystÄ™pujÄ…cych wartoÅ›ci
    wybrane_wartosci = posortowane_wartosci[:3]
    
    # Dodawanie wybranych wartoÅ›ci do sÅ‚ownika wynikowego
    wartosci_wystapien[przedzial] = wybrane_wartosci

# WyÅ›wietlanie wybranych wartoÅ›ci
for przedzial, wybrane_wartosci in wartosci_wystapien.items():
    print(f"PrzedziaÅ‚ {przedzial}:")
    for wartosc, wystapienia in wybrane_wartosci:
        print(f" - WartoÅ›Ä‡: {wartosc}, wystÄ…pienia: {wystapienia}")
```
