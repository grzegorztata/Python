```python
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
import pandas as pd
```

```python
df = pd.DataFrame({'A':np.random.randn(200), #'standard normal distribution'
                 'B':[dt.datetime(2019,1,1)+dt.timedelta(days=x) for x in range(200)], # czas od 2019-01-01 i inkrementuje o 1 dzień
                  'C':np.arange(1,201)})
df
```

```python
## Tworzymy histogram dla wartości zawartych w kolumnie 'A'
```

```python
df['A'].hist(bins=30) # dzielimy wartości na 30 przedziałów
```

```python
# Wykres kolumnowy dla kolumny 'C'
```

```python
df['C'].loc[:10].plot(kind='bar')
```

```python
## Rysowanie konkretnego rodzaju wykresu
```

```python
df.plot.line(x='B',y='A',figsize=(12,5), c='red',lw=2)
```
