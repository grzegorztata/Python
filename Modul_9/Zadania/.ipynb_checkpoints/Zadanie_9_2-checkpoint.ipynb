```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
df = pd.read_csv('Files/HRDataset.csv')
df.dropna(thresh=2,inplace=True)  # thresh - W tym przypadku, jeśli wiersz zawiera co najmniej 2 niebrakujące wartości, to nie zostanie usunięty
df.head()
```

### Ad.1 - Czy istnieje zależność pomiędzy tym, kto jest bezpośrednim przełożonym (ManagerName, ManagerID) danego pracownika, a oceną wydajności pracy (PerformanceScore)?

```python
# Zmieniamy typ danych na String (błąd unsupported operand type(s) for -: 'str' and 'str')
df['ManagerID'] = df['ManagerID'].fillna('999')
df['ManagerID'] = df['ManagerID'].astype(int)

counts = df.groupby(['ManagerID', 'PerformanceScore']).size().reset_index(name='Counts')

plt.figure(figsize=(10,5))
sns.barplot(x='ManagerID', y='Counts', hue='PerformanceScore', data=counts)

plt.xlabel('Manager ID')
plt.ylabel('Ilość')
plt.title('Porównanie ManagerID i PerformanceScore')
plt.show()
```

###### Z przedstawionego wykresu można zauważyć, że Manager ID 14 opiekuje się największą liczbą pracowników z najlepszą oceną. Pracownicy tego managera z najlepszą oceną nie wynoszą jednak 100%.  100% efektywność mają pracownicy managerów ID: 1, 9 i 14, ale próba pracowników jest zbyt mała

###### Aby to przeliczyć 'PerformanceScore' do każdej oceny przypiszę wartości dla każdej oceny

```python
df_performance_data = df[['Employee_Name','ManagerID', 'ManagerName', 'PerformanceScore']].copy()  # dodano kopiowanie aby utworzyć pustą kolumnę
df_performance_data.loc[df_performance_data['PerformanceScore'] == 'Fully Meets', 'ScoreValue'] = 4
df_performance_data.loc[df_performance_data['PerformanceScore'] == 'Exceeds', 'ScoreValue'] = 3
df_performance_data.loc[df_performance_data['PerformanceScore'] == 'Needs Improvement', 'ScoreValue'] = 2
df_performance_data.loc[df_performance_data['PerformanceScore'] == 'PIP', 'ScoreValue'] = 1
df_performance_data.head()
```

###### Sprawdzam ilu pracowników ma każdy manager

```python
manager_counts = df_performance_data['ManagerID'].value_counts().reset_index()
manager_counts.columns = ['ManagerID', 'Count']
manager_counts
```

###### Usunę użytkowników nieprzypisanych do żadnego managera, czyli tych co są przypisani do 999

```python
manager_counts = manager_counts.drop(manager_counts[manager_counts['ManagerID'] == 999].index)  # ManagerID (int)
```

###### Zostawię 50% Managerów z największą liczbą przypisanych pracowników

```python
total_managers = len(manager_counts)
managers_to_remove = int(total_managers * 0.5)
representative_managers = manager_counts.nlargest(total_managers - managers_to_remove, 'Count').index
filtered_managers = manager_counts.loc[representative_managers]
filtered_managers
```

###### Mergujemy tabele po kolumnie 'ManagerID'

```python
df_half_managers = pd.merge(df_performance_data, filtered_managers, on='ManagerID')
df_half_managers = df_half_managers[['Employee_Name', 'ManagerID', 'ManagerName', 'PerformanceScore', 'ScoreValue']]
df_half_managers.head()  # Z 310 pozostało 235 Pracowników
```

###### Pogrupowałem kolumny 'ManagerID' i przeliczyłem średnią ocen dla każdego Managera. Następnie utworzyłem nową tabelę ze średnią oceną Pracowników przypisanych do Managera

```python
# Tutaj mi ktoś pomógł, bo nie wiedziałem jak utworzyć tabelę ze średnią
total_score_value = pd.DataFrame(columns=['ManagerID', 'ManagerName', 'AverageScore'])

grouped = df_half_managers.groupby(['ManagerID', 'ManagerName'])
for (manager_id, manager_name), group in grouped:
    average_score = group['ScoreValue'].mean()
    total_score_value = pd.concat([total_score_value, pd.DataFrame({'ManagerID': manager_id, 'ManagerName': manager_name, 'AverageScore': average_score}, index=[0])])
    # konkatenacja wewnątrz pętli, a następnie dodaje wiersze
total_score_value
```

```python
total_score_value_sorted = total_score_value.sort_values('AverageScore', ascending=False)
total_score_value_sorted
```

```python
first_manager_name = total_score_value_sorted['ManagerName'].iloc[0]
score_percent = (total_score_value_sorted['AverageScore'].iloc[0].round(2)) / 4 * 100
```

```python
print(f'''W mojej opinii nie istnieje żadna zależność pomiędzy Managerem i Oceną, 
ale ustaliłem że najlepszym Managerem jest {first_manager_name} z efektywnością {score_percent}%''')
```

# Ad.2 - Jakie źródła pozyskania pracownika (Recruitment Source) są najlepsze, jeśli zależy nam na jak najdłuższym stażu pracowników?

```python
df['DOB'] = pd.to_datetime(df['DOB'],format='%m/%d/%y')
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'],format='%m/%d/%y')
df['DateofHire'] = pd.to_datetime(df['DateofHire'],format='%m/%d/%Y')
df[['DateofTermination','DOB','DateofHire','RecruitmentSource']].head()
```

```python
def count_seniority(row):
    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2019,9,27)
    else:
        end_date = row['DateofTermination']
    
    return (end_date - row['DateofHire'])/np.timedelta64(1,'Y')
```

```python
df['Seniority'] = df.apply(lambda row: count_seniority(row),axis=1)
```

```python
df[['RecruitmentSource','Seniority']]
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.boxplot(x='RecruitmentSource',y='Seniority',data=df)
```

###### Wykres pokazuje, że jeżeli zależy nam na jak najdłuższym stażu pracownika, to najlepszym źródłem jest OTHERS, ponieważ górny wąs jest na jednym z najwyższych poziomów i mediana Źródła jest najwyżej od pozostałych. Problemem tutaj może być liczba zatrudnionych w ten właśnie sposób, więc ja bym sprawdził liczbę zatrudnionych według źródła i usunął te, które mają najmniejszą sumę

```python
recruitment_source_counts = df['RecruitmentSource'].value_counts().reset_index()
recruitment_source_counts.columns = ['RecruitmentSource', 'Count']
recruitment_source_counts
```

###### Other było użyte przy zatrudnianiu 9 Pracowników. Samo źródło tak naprawdę niewiele nam mówi, więc usunę to źródło oraz te, które były użyte tylko 1 raz

```python
recruitment_source_counts = recruitment_source_counts.drop(recruitment_source_counts[recruitment_source_counts['RecruitmentSource'] == 'Other'].index)
recruitment_source_counts = recruitment_source_counts.drop(recruitment_source_counts[recruitment_source_counts['Count'] == 1].index)
recruitment_source_counts
```

```python
df_for_recruitment_src = pd.merge(recruitment_source_counts, df, on='RecruitmentSource')
df_for_recruitment_src.head()
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.boxplot(x='RecruitmentSource',y='Seniority',data=df_for_recruitment_src)
```

### Po usunięciu zbędnych żródeł uważam, że najlepszym źródłem pozyskania Pracowników jest Billboard głównie ze względu na wysoko umiejscowioną medianę oraz jeden z najwyżej umiejscowionych 3 kwartyl
