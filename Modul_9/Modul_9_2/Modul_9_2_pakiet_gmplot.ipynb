```python
import pandas as pd
import gmplot
import numpy as np
from IPython.display import IFrame
```

```python
zip_codes = pd.read_csv('Files/us-zip-code-latitude-and-longitude.csv',sep=';')
zip_codes.head()
```

```python
df = pd.read_csv('Files/HRDataset.csv')
df.head()
```

##### Kopiujemy tabelÄ™ i wybieramy kolumny 'Zip' i 'PayRate'

```python
df_mapping = df[['Zip','PayRate']].copy()
df_mapping.head()
```

```python
df_mapping = df_mapping.join(zip_codes.set_index('Zip')[['Latitude','Longitude']],on='Zip',how='left')
df_mapping.head()
```

```python
locations = df_mapping[['Latitude','Longitude']].dropna()
weights = df_mapping['PayRate'].dropna()
latitude = locations['Latitude']
longitude = locations['Longitude']
weights /= weights.max()
```
