```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
bike_data = pd.read_csv('Files/daily-bike-share.csv', index_col='instant')
display(bike_data)
```

#### ZwiÄ…zek pogody do liczby wyporzyczonych rowerÃ³w

```python
sns.pairplot(bike_data[['temp','hum','windspeed','rentals']])
plt.show()
```

```python
#### Mapowanie obserwacji w zaleÅ¼noÅ›ci od kategorii
```

```python
sns.pairplot(bike_data[['season','temp','hum','windspeed','rentals']], hue='season', palette='bright')
plt.show()
```

### Korelacja Pearsona

```python
plt.figure(figsize=(9,6))
sns.heatmap(bike_data[['temp', 'hum', 'windspeed', 'rentals']].corr(), vmin=-1, vmax=1, annot=True)
plt.show()
```
