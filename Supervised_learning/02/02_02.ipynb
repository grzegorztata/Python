```python
#import all libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder 
```

```python
student1 = pd.read_csv("../Files/student-mat.csv", sep = ';', header = 0)
student2 = pd.read_csv("../Files/student-por.csv", sep = ';', header = 0)
```

```python
student1.head()
```

```python
# View first few rows of data
student2.head()
```

```python
students = pd.concat([student1,student2])
```

```python
students.head()
```

```python
# Print shape of dataset
print(students.shape)
```

```python
# Check for missing data
total = students.isnull().sum().sort_values(ascending=False)
# Display total
print(total)
```

```python
# Describe dataset
students.describe()
```

```python
# Wybierz tylko kolumny numeryczne
numeric_columns = students.select_dtypes(include=['float64', 'int64'])
# Oblicz korelację dla wybranych kolumn numerycznych
students_matrix = numeric_columns.corr()
```

```python
# Inspect feature correlations
plt.figure(figsize=(12, 8))
ax = sns.heatmap(students_matrix, cmap='Greens', annot=True)
plt.show()
```

```python
(students.shape)
```

```python
students_matrix[['G3']].sort_values(by='G3', ascending = False)
```

```python
# One hot encode labels
encoded_students = OneHotEncoder().fit_transform(students).toarray()
print(encoded_students)
```

```python
(encoded_students.shape)
```

```python
# Save encoded_students as pandas DataFrame
students_df = pd.DataFrame(encoded_students)
```

```python
g = sns.PairGrid(students[['G2','G1','Medu','studytime', 'Fedu']])
g = g.map_upper(sns.kdeplot, cmap='YlGnBu', fill=True, thresh=False)
g = g.map_diag(plt.hist, color='#6cbd5e')
g = g.map_lower(sns.scatterplot, color='#b8e0b1', edgecolor='k')

g.fig.tight_layout()
g.fig.suptitle('Relationship')
plt.subplots_adjust(top=0.9)
```
